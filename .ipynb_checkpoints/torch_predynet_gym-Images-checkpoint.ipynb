```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F

from helpers import *

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from jupyterthemes import jtplot

jtplot.style(theme='oceans16')
# torch.manual_seed(1)

import gym
from scipy import stats
```

```python
env = gym.make('LunarLander-v2')
# env = gym.make('MountainCar-v0')
# env = gym.make('BipedalWalker-v2')
# env = gym.make('CarRacing-v0')
```

```python
env.reset()
action = env.action_space.sample()
observation = env.step(action)
obs_space = observation[0].shape[0]
```

```python
def gauss_act(x):
    return (torch.exp(-(torch.pow(x,2))/2))

def gauss_diff(x):
    return torch.mul(x,torch.exp(-(torch.pow(x,2))/2))
```

```python
inSize = env.observation_space.shape[0]
timelength = 10000
try:
    act_space = env.action_space.n
except:
    act_space = env.action_space.shape[0]
    
ySize = act_space
```

```python
# Torchus

y_pred = Variable(torch.randn(ySize).clamp(-1,1))
er = Variable(torch.zeros(inSize))

w_mu = 0
w_std = 0.1

W = Variable(torch.FloatTensor(stats.truncnorm.rvs(w_mu, w_std, size=[inSize,ySize])),
             requires_grad=True)

V = Variable(torch.FloatTensor(stats.truncnorm.rvs(w_mu, w_std, size=[ySize,inSize])),
             requires_grad=True)

R = Variable(torch.FloatTensor(stats.truncnorm.rvs(w_mu, w_std, size=[ySize,ySize])),
            requires_grad=True)

# Forward weights

WA = Variable(torch.FloatTensor(stats.truncnorm.rvs(w_mu, w_std, size=[act_space,ySize])),
             requires_grad=True)
```

```python
lr = 0.1
lrec = 0.01

optimizer = torch.optim.SGD([W,V],lr=lr)
criterion = nn.MSELoss()
# criterion = F.smooth_l1_loss

Roptimizer = torch.optim.SGD([R],lr=lrec)
Rcriterion = F.smooth_l1_loss

Aoptimizer = torch.optim.SGD([WA], lr=lr)
Acriterion = F.smooth_l1_loss

def er_loss(x,y):
    return torch.mean(torch.pow(x-y, 2))
```

```python
E = np.zeros(timelength)
Y = np.zeros([ySize,timelength]) 
RE = np.zeros(timelength)
AC = np.zeros(timelength)
Rew = np.zeros(timelength)

yprev = y_pred.unsqueeze(0)
y = yprev
cnt = 0
env.reset()
action = env.action_space.sample()

for t in range(timelength):
        
    if np.random.rand() > 0.5:
        action = env.action_space.sample()
        
    try:
        A = Variable(torch.FloatTensor(np.eye(act_space)[action]))
    except:
        A = Variable(torch.FloatTensor(action))
        
        
    ObsOut = env.step(action)
    try:
        observation = Variable(torch.FloatTensor((ObsOut[0]).flatten()))
    except:
        observation = Variable(torch.FloatTensor(np.float32(ObsOut[0]).flatten()))

    done = ObsOut[2]
    Rew[t] = ObsOut[1]
    
    optimizer.zero_grad()
    Roptimizer.zero_grad()
    Aoptimizer.zero_grad()

#     for ind in range(2):
    er = (torch.mm(yprev,V) - observation)

    y = F.tanh(torch.mm(er.detach(), W) + torch.mm(yprev.detach(), R)
              + torch.mm(A.unsqueeze(0),WA))


    loss = criterion(torch.mm(yprev,V),observation)
    rloss = Rcriterion(torch.mm(er.detach(), W.detach()), torch.mm(yprev.detach(), R))
    E[t] = loss.data.numpy()
    Y[:,t]  = y.data.numpy()
    RE[t] = rloss.data.numpy()
#     AC[t] = action
    
    loss.backward()
    rloss.backward()
    optimizer.step()
    Roptimizer.step()
    Aoptimizer.step()
    
    yprev = y
    
    if done:
        env.reset()
        cnt +=1

plt.subplot(211)
plt.plot(E)
plt.subplot(212)
plt.plot(RE)
```

```python
testlength = 1000
Rec = np.zeros([len(ObsOut[0]), testlength])
RealObs = np.zeros_like(Rec)
env.reset()
for t in range(testlength):
        
    if np.random.rand() > 0.5:
        action = env.action_space.sample()
        
    try:
        A = Variable(torch.FloatTensor(np.eye(act_space)[action]))
    except:
        A = Variable(torch.FloatTensor(action))
        
    ObsOut = env.step(action)
    observation = Variable(torch.FloatTensor(ObsOut[0]))
    done = ObsOut[2]
    Rew[t] = ObsOut[1]

#     for ind in range(13):
    er = (torch.mm(yprev,V) - observation)

    y = F.tanh(torch.mm(er.detach(), W) + torch.mm(yprev.detach(), R)
              + torch.mm(A.unsqueeze(0),WA))

        
    tmp = (torch.mm(yprev,V))
    Rec[:,t] = tmp.data.numpy()
    RealObs[:,t] = observation.data.numpy()  
    
    yprev = y

    if done:
        env.reset()
     
plt.subplot(211)
plt.plot(Rec.T);
plt.subplot(212)
plt.plot(RealObs.T);
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F

from helpers import *

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from jupyterthemes import jtplot

jtplot.style(theme='oceans16')
# torch.manual_seed(1)

import gym
```

```python
env = gym.make('CartPole-v0')
# env = gym.make('LunarLander-v2')
```

```python
env.reset()
action = env.action_space.sample()
observation = env.step(action)
observation
```

```python
env.observation_space.shape
```

```python
ySize = 64
inSize = env.observation_space.shape[0]
timelength = 12000
act_space = env.action_space.n
```

```python
# Torchus

y_pred = Variable(torch.randn(ySize).clamp(-1,1))
er = Variable(torch.zeros(inSize))

# Initiliaze weights
W = Variable(torch.randn(inSize,ySize), requires_grad=True)
W.clamp(-0.1,0.1);
V = Variable(torch.randn(ySize,inSize), requires_grad=True)
V.clamp(-0.1,0.1);

R = Variable(torch.randn(ySize,ySize),requires_grad=True)
R.clamp(-0.1,0.1);

# Forward weights

WA = Variable(torch.randn(act_space,ySize),requires_grad=True)
WA.clamp(-0.1,0.1);
```

```python
lr = 0.1
lrec = 0.01

optimizer = torch.optim.RMSprop([W,V],lr=lr)
criterion = nn.MSELoss()

Roptimizer = torch.optim.RMSprop([R],lr=lrec)
Rcriterion = F.smooth_l1_loss

Aoptimizer = torch.optim.RMSprop([WA], lr=lr)
Acriterion = F.smooth_l1_loss

def er_loss(x,y):
    return torch.mean(torch.pow(x-y, 2))
```

```python
E = np.zeros(timelength)
Y = np.zeros([ySize,timelength]) 
RE = np.zeros(timelength)
AC = np.zeros(timelength)
Rew = np.zeros(timelength)

yprev = y_pred.unsqueeze(0)
y = yprev
cnt = 0
env.reset()
action = env.action_space.sample()

for t in range(timelength):
        
    if np.random.rand() > 0.8:
        action = env.action_space.sample()
        
    A = Variable(torch.FloatTensor(np.eye(act_space)[action]))
    ObsOut = env.step(action)
    observation = Variable(torch.FloatTensor(ObsOut[0]))
    done = ObsOut[2]
    Rew[t] = ObsOut[1]
    
#     for ind in range(4):
    optimizer.zero_grad()
    Roptimizer.zero_grad()
    Aoptimizer.zero_grad()

    for ind in range(4):
        er = torch.mm(yprev,V) - observation

        y = F.tanh(torch.mm(er.detach(), W) + torch.mm(yprev.detach(), R)
                  + torch.mm(A.unsqueeze(0),WA))

        yprev = y


    loss = criterion(torch.mm(yprev,V),observation)
    rloss = Rcriterion(torch.mm(er.detach(), W.detach()), torch.mm(yprev.detach(), R))
    E[t] = loss.data.numpy()
    Y[:,t]  = y.data.numpy()
    RE[t] = rloss.data.numpy()
    AC[t] = action
    
    loss.backward()
    rloss.backward()
    optimizer.step()
    Roptimizer.step()
    Aoptimizer.step()

    AC[t] = done
    if done:
        env.reset()
        cnt +=1

plt.subplot(211)
plt.plot(E)
plt.subplot(212)
plt.plot(RE)
```
